```python
import matplotlib.pyplot as plt  
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
# data
x = range(-100,100)
y= [(x**2-x +2) for x in x]

# Plot x and y
plt.plot(x,y)

#add annotation
plt.annotate('origin', xy=(0,0), arrowprops = {}, xytext = (0,2000))
```

```python
x1 = range(0,100)
y1 = np.sqrt(x1)

#plot
plt.plot(x1,y1)

#additions
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
```

```python
x2 = range(0,100)
y2 = np.power(x2,3)

#plt
plt.plot(x2,y2)

# additions
plt.title('$ y = x^3 $')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
```

```python
x3 = [x* 0.1 for x in range(0,100)] 
y3 = [2**n for n in x3]

#plot
plt.plot(x3,y3)

#additions
plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$x^2$')
```

```python
x4 = [x * 0.1 for x in range (0,100)]
y4 = [1/(n+1) for n in x4]

# plot
plt.plot(x4,y4)

#additions
plt.title('$y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
```

```python
plt.figure(figsize = (10,10))

# plot 1
plt.subplot(221)
#plot
plt.plot(x1,y1)

#additions
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')

# plot 2
plt.subplot(222)
plt.plot(x2,y2)

# additions
plt.title('$ y = x^3 $')
plt.xlabel('$x$')
plt.ylabel('$x^3$')

#plot 3
plt.subplot(223)
plt.plot(x3,y3)

#additions
plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$x^2$')

#plot 4
plt.subplot(224)
plt.plot(x4,y4)

#additions
plt.title('$y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')
```
