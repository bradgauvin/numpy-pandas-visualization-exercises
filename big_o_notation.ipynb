```python
import math
import matplotlib.pyplot as plt

#x values

x = range(1,20)

#y values
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = x
y4 = [n * math.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]
```

```python
plt.figure(figsize= (15,7))
plt.title('Big O Notation', fontsize = 15)
plt.xlabel('Elements', fontsize = 14)
plt.ylabel('Operations', fontsize = 14)

# Plots:
plt.plot(x,y1, label = '$\mathcal{O}{n+1}$', ls = '--')
plt.plot(x,y2, label = '$\mathcal{O}(log(n))$', ls = ':')
plt.plot(x,y3, label = '$\mathcal{O}(n)$', ls = '-.')
plt.plot(x,y4, label = '$\mathcal{O}(n*log(n))$', ls = '-')
plt.plot(x,y5, label = '$\mathcal{O}(n^2)$', ls = '--')
plt.plot(x,y6, label = '$\mathcal{O}(2^{n})$', ls = ':')
plt.plot(x,y7, label = '$\mathcal{O}(n!)$', ls = "-.")
plt.plot(x,y8, label = '$\mathcal{O}(n^n)$', ls = '-', marker = 'o')

#Additions
plt.xlim(1,7)
plt.ylim(0,40)
plt.legend(fontsize = 15)
```
